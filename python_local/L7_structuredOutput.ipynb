## Structured Output Example

```python
from typing_extensions import TypedDict

# Define the expected response format using a TypedDict
class ContactInfo(TypedDict):
    """Contact information for a person."""
    name: str # The name of the person
    email: str # The email address of the person
    phone: str # The phone number of the person
```

```python
from langchain_ollama import ChatOllama

# initialize the local ollama language model (here, gpt-oss) with desired
# parameters
model = ChatOllama(
    model="gpt-oss",
    temperature=0,
)
```

Since we use a local model using Ollama (here, gpt-oss) that may not support
native structured output, we can use the `response_format` parameter of
`create_agent` to specify the expected output format. For such models, we can
use a tool calling strategy, using `ToolStrategy`, to obtain the structured
output. For further details see the documentation for `ToolStrategy`
[here](https://docs.langchain.com/oss/python/langchain/structured-output#tool-calling-strategy)

```python
from langchain.agents import create_agent
from langchain.agents.structured_output import ToolStrategy

# Create your agent! Add the model object you just created, a prompt etc.
agent = create_agent(model=model, response_format=ToolStrategy(ContactInfo))
```

```python
recorded_conversation = """We talked with John Doe. He works over at Example. His number is, let's see, 
five, five, five, one two three, four, five, six seven. Did you get that?
And, his email was john at example.com. He wanted to order 50 boxes of Captain Crunch."""

result = agent.invoke(
    {"messages": [{"role": "user", "content": recorded_conversation}]}
)

result["structured_response"]
```

#### Multiple data types are supported 

* pydantic `BaseModel`
* `TypedDict`
* `dataclasses`
* json schema (dict)

```python
from pydantic import BaseModel


class ContactInfo(BaseModel):
    name: str
    email: str
    phone: str


agent = create_agent(model=model, response_format=ToolStrategy(ContactInfo))

recorded_conversation = """ We talked with John Doe. He works over at Example. His number is, let's see, 
five, five, five, one two three, four, five, six seven. Did you get that?
And, his email was john at example.com. He wanted to order 50 boxes of Captain Crunch."""

result = agent.invoke(
    {"messages": [{"role": "user", "content": recorded_conversation}]}
)

result["structured_response"]
```
